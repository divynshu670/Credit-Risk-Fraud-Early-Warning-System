```python
import pandas as pd

features = pd.read_csv("../data/features.csv")

# Normalize helper
def normalize(col):
    return (col - col.min()) / (col.max() - col.min() + 1e-6)

features["risk_score"] = (
    0.35 * normalize(features["night_txn_ratio"]) +
    0.30 * normalize(features["default_count"]) +
    0.20 * normalize(features["avg_delay"]) +
    0.15 * normalize(features["avg_amount"])
)

def segment(score):
    if score >= 0.7:
        return "High"
    elif score >= 0.4:
        return "Medium"
    return "Low"

features["risk_segment"] = features["risk_score"].apply(segment)

features.to_csv("../data/risk_scored_users.csv", index=False)
```
