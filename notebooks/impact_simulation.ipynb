```python
import pandas as pd

users = pd.read_csv("../data/risk_scored_users.csv")
repayments = pd.read_csv("../data/repayments.csv")

baseline_defaults = repayments[repayments["repayment_status"] == "default"].shape[0]

high_risk_users = users[users["risk_segment"] == "High"]["user_id"]

post_policy_defaults = repayments[
    (repayments["repayment_status"] == "default") &
    (~repayments["user_id"].isin(high_risk_users))
].shape[0]

impact = {
    "baseline_defaults": baseline_defaults,
    "post_policy_defaults": post_policy_defaults,
    "reduction_pct": (baseline_defaults - post_policy_defaults) / baseline_defaults
}

impact
```
